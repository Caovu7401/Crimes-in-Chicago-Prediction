```python
#Số lượng dữ liệu bị Null là khá ít so với toàn bộ, ta có thể xóa đi hết
data = data.dropna()
```

***Ở column 'Case Number', dễ thấy có khá nhiều dòng mang chung một case, điều này là vì một người có thể vi phạm nhiều tội cùng lúc. Ta sẽ cân nhắc để xóa các Case Number trùng lặp, chỉ cần giữ lại Case
 có Updated On mới nhất***

```python
#Chuyển đổi column Updated On	sang dạng datetime
data.loc[:, 'Updated On'] = pd.to_datetime(data['Updated On'], format='%Y-%m-%d %H:%M:%S')
```

```python
# Sắp xếp data theo 'Case Number' và 'Update on'
data = data.sort_values(by=['Case Number', 'Updated On'], ascending=[True, False])

# Lọc bỏ các giá trị trùng lặp của Case Number, chỉ giữ lại giá trị mới nhất
data = data.drop_duplicates(subset=['Case Number'], keep='first')
```

# **SAMPLING DATA**

***Dựa theo Primary Type, ta thấy có 4 giá trị chiếm đa số là THEFT, BATTERY, CRIMINAL DAMAGE, NARCOTICS. Ta sẽ chọn ra 4 giá trị này. Tuy nhiên, có một số loại tội phạm cũng tương tự như 4 loại trên. Vì vậy. Ta sẽ đổi tên một số loại tội danh và sau đó mới chọn ra 4 tội***

```python
#THEFT vs MOTOR VEHICLE THEFT-Đổi tên 'MOTOR VEHICLE THEFT' thành 'THEFT'
data.loc[data['Primary Type'] == 'MOTOR VEHICLE THEFT', 'Primary Type'] = 'THEFT'

#Tương tự với các cặp còn lại
data.loc[data['Primary Type'] == 'ASSAULT', 'Primary Type'] = 'BATTERY'
data.loc[data['Primary Type'] == 'ARSON', 'Primary Type'] = 'CRIMINAL DAMAGE'
data.loc[data['Primary Type'] == 'OTHER NARCOTIC VIOLATION', 'Primary Type'] = 'NARCOTICS'
```

```python
data['Primary Type'].value_counts()
```

```python
#Chọn ra 4 loại dữ liệu lớn nhất:
data = data.loc[data['Primary Type'].isin(['THEFT', 'BATTERY', 'CRIMINAL DAMAGE', 'NARCOTICS'])]
```

# **DROPPING OULIERS**

```python
#Xóa đi một số columns không cần thiết (ở đây ta thấy giá trị tọa độ X, Y rất lớn, vượt xa qua giá trị của các columns trên):
#Thay vào đó, nên sử dụng Latitude, Longitude vì có giá trị phù hợp hơn
data = data.drop(columns = ['X Coordinate', 'Y Coordinate', 'Location'])
```
